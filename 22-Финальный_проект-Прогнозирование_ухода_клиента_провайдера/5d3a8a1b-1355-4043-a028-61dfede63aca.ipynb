#### Заполнение пропусков в столбце 'EndDate' и добавление новых параметров

```python
end_date = pd.DataFrame({'before': df.EndDate.value_counts()})
```

```python
def add_month(date1, month_to_add):
    year = date1.year
    month = date1.month + month_to_add
    if ((month % 12) == 0) and ((month_to_add % 12) == 0): month -= 1
    if month>12:
        year += month // 12
        month = month % 12
    day= date1.day
    return pd.Timestamp(year, month, day)

def end_of_multiple_period(row):
    end = row.BeginDate
    if pd.isna(row.EndDate): end_date = pd.Timestamp(2020, 2, 1)
    else: end_date = row.EndDate
    while end < end_date:
        end = add_month(end, row.Type)
    return end

df['end_of_multiple_period'] = df.apply(end_of_multiple_period, axis=1)

def restore_end_date(row):
    if not pd.isna(row.EndDate): return row.EndDate
    return row.end_of_multiple_period

df.EndDate = df.apply(restore_end_date, axis=1)

df['d_time'] = (df.EndDate - df.BeginDate).apply(lambda x: x.days)
df['paid_periods'] = round( ((df.EndDate - df.BeginDate).apply(lambda x: x.days))*12/365,0) / df.Type
```

```python
times_rows = ['BeginDate', 'EndDate', 'end_of_multiple_period']
df1 = df.drop(times_rows, axis=1).copy()

train1, valid1 = train_test_split(df1, test_size=TEST_SIZE, random_state=RANDOM_STATE)

def get_preprocessor(*, model, num, cat):
    encoder_lin = OneHotEncoder(handle_unknown='ignore', sparse_output=False, drop='first')
    # handle_unknown='ignore' - бесперебойная работа при появлении в новых
    # данных ранее неизвестных значений признаков. В итоговую матрицу в итоге не войдут
    encoder_tree = OrdinalEncoder(handle_unknown='use_encoded_value', unknown_value=-1)
    # handle_unknown='use_encoded_value' - ранее неизвестные значения кодируются константным значением
    # unknown_value - само значение-константа
    if model == LogisticRegression: encoder = encoder_lin
    else: encoder = encoder_tree
    num_transformer = Pipeline(steps=[('scaler', StandardScaler()),
                                  ('imputer', KNNImputer())])
    cat_transformer = Pipeline(steps=[('encoder', encoder), 
                                  ('scaler', StandardScaler()),
                                 ('imputer', KNNImputer())])
    return ColumnTransformer(transformers=[
    ('num', num_transformer, num),
    ('cat', cat_transformer, cat)
    ])
```

```python
try:
    result1 = pd.read_csv('3c16-1.csv')
    # не работает после восстановления из файла. Проще перегенерировать с 1 набором параметров
    #result1.feature_importances = pd.DataFrame(result1.feature_importances)
except:
    result1 = pd.DataFrame(columns=['label', 'best_param', 'time_fit', 'time_predict', 'roc_auc',
                                'best_model', 'param_grid', 'feature_importances', 'total_time'])
result1
```

```python
def get_metrics(*, param_grid, model, label, train, cv=7, verbose=3,
                num=IS_NUMERIC, cat=IS_CATEGORIAL):
    preprocessor = get_preprocessor(model=model, num=num, cat=cat)
    if model != DummyClassifier:
        pipeline_current = Pipeline(steps = [('preprocessor', preprocessor),
                                         ('regressor', model(random_state=RANDOM_STATE) )])
    else:
        pipeline_current = Pipeline(steps = [('preprocessor', preprocessor),
                                         ('regressor', model() )])
    model_current = GridSearchCV(pipeline_current, verbose=verbose, param_grid=param_grid, cv=cv,
        refit=True, error_score = 'raise', scoring='roc_auc')
    if verbose >0: display(model_current)
    model_current.fit(train.drop('escape', axis=1), train.escape)
    time_fit = model_current.refit_time_
    exist_hyper=[]
    for i in list(model_current.get_params().keys()):
        if 'estimator__regressor__' in i: exist_hyper.append(i.replace('estimator__regressor__', ''))
    if verbose >2 : display(f"доступные гиперпараметры: {exist_hyper}")
    display(f'Лучшие гиперпараметры: {model_current.best_params_}')
    display(f"Достигнуто roc-auc: {model_current.best_score_}")
    start=time.time()
    tmp = model_current.best_estimator_.predict(train.drop('escape', axis=1))
    time_predict = time.time() - start
    if model in [RandomForestClassifier, CatBoostClassifier]:
        feature_importances = pd.DataFrame(
            data={'score': model_current.best_estimator_._final_estimator.feature_importances_},
            index=model_current.best_estimator_['preprocessor'].get_feature_names_out()
                ).sort_values(by='score')
    if model == LogisticRegression:
        feature_importances = pd.DataFrame(
            data={'score': model_current.best_estimator_._final_estimator.coef_[0]},
            index=model_current.best_estimator_['preprocessor'].get_feature_names_out()
            ).sort_values(by='score')
    if model != DummyClassifier:
        if verbose > 0: feature_importances.plot(kind='barh', grid=True,figsize=(18,10), legend=False);
    else: feature_importances = None
    return pd.Series({'label':label, 'best_param': model_current.best_params_, 'time_fit': time_fit,
        'time_predict': time_predict, 'best_model': model_current.best_estimator_,
        'param_grid':param_grid, 'total_time': None,
        'feature_importances': feature_importances, # 
        'roc_auc': model_current.best_score_})
```

```python
def get_metrics_cached(*, param_grid, model, label, train, cv=7, verbose=3,
                num=IS_NUMERIC, cat=IS_CATEGORIAL, file = '3c16-1.csv', ro=False):
    global result1
    if len(result1[result1.label == label]) == 0:
        time_start = time.time()
        _ = get_metrics(param_grid=param_grid, model=model, label=label, train=train, cv=cv,
                             verbose=verbose, num=num, cat=cat)
        _.total_time = time.time() - time_start
        result1.loc[len(result1)] = _
        if not ro: result1.to_csv(file, index=False)
    else:
        print('В таблице результатов уже есть запись с такой меткой.\
 Будут использоваться уже предвычесленные результаты')
        display(f"best_param: {result1[result1.label == label].iloc[-1].best_param}")
        display(f"best_roc_auc: {result1[result1.label == label].iloc[-1].roc_auc}")
        display(f"Затраченное время проверки: {result1[result1.label == label].iloc[-1].total_time / 60:.1f} минут")
```

         * столбец `gender`. Содержал 2 уникальных значения: \[`Female`, `Male`\]. Было принято решение о смене имени столбца на `is_male` и замене на значения \[0, 1\] соответственно. Это позволило преобразовать столбец к целому типу.
         * столбец `InternetService`. Содержал значения \[`DSL`, `Fiber optic`, nan\]. Аналогично было заменено имя на `is_optic` и столбец был преобразован к целому типу с поддержкой NaN.
         * столбец `Type`. Содержал значения \[`Month-to-month`, `One year`, `Two year`\]. Единственный столбец, которые был преобразован из категориального в числовой, со значениями \[1, 12, 24\]. Тип столбца был заменен на целый тип. В дальнейшем именно численные значения использовались для вычисления значений вновь созданных признаков.
         * столбец `PaymentMethod`. Содержал значения \[`Electronic check`, `Mailed check`, `Bank transfer (automatic)`, `Credit card (automatic)`\]. Это категориальный тип данных, но для построения парных графиков был перекодирован в автоматическом режиме в целый тип (seaborn.pairplot ругается на тип object). Для возможности вернуть его обратно был сохранен словарь обратного преобразования. Для обучения модели в дальнейшем, данные столбца все равно будут пропушены через енкодер, а ему без разницы, текст там будет или число-категория.
         * у оставшихся столбцов с бинарными данными был изменен тип на целое, либо целое с поддержкой NaN, соответственно.
      * добавление новых признаков:
         * `end_of_multiple_period` - дата конца контракта, если бы от начала контракта прошло целое количество периодов оплаты (1, 12 или 24 месяца). Промежуточный признак, облегчающий вычисление других.
         * `d_time` - время между началом и окончанием контракта в днях (вычисляется после заполнения пропусков в столбце `EndDate`)
         * `paid_periods` - время между началом и окончанием контракта в периодах оплаты (которые равны 1, 12 или 24 месяцам). Дробное значение этого параметра - индикатор того, что пользователь менял тариф в сторону увеличения.
         * в процессе работы создавались еще столбцы "окончание первого периода" и "пользователь ушел после первого периода", но в итоге от их использования отказался, т.к. они не подтвердили своей ценности
      * восстановление пропусков в столбце `EndDate`: пропуски заполнялись значением поля `end_of_multiple_period` (время истечения целого количества контрактов с момента заключения, ближайшее к дате 2020-02-01 справа)
      * удаление противоречивой информации: если `paid_periods` меньше 1, то это значит, что между началом и окончанием контракта прошло время меньше оплаченного, что непонятно как интерпретировать. Было решено такие данные считать ошибочными. Было удалено 2% данных.
      * предварительное исследование данных:

         * Всевозможные опции интернета и телефонии практически не оказывают влияние на вероятность ухода клиента
* Обучение моделей машинного обучения:
   * созданна подпрограмма генерирования последовательности предобработки, в зависимости от типа модели (энкодер `OneHotEncoder` для линейной модели и `OrdinalEncoder` для остальных)
   * создана подпрограмма для получения всех необходимых метрик, на основе выбранной модели и гиперпараметров
   * создана подпрограмма, сохраняющая результаты расчетов оптимальных гиперпараметров в файл, с последующей загрузкой. На этом этапе наступил на грабли, что из csv файла не получается сходу считать вложенный датафрейм/модель/паплайн. Решил костыльным методом: в принципе достаточно информации о оптимальных гиперпараметрах (о них информация сохраняется/загружается корректно), после чего запускается `GridSearchCV` точечно с этими гиперпараметрами (что позволяет обращаться к массиву результатов замеров унифицированным способом). Это позволило сэкономить массу машиночасов (повторно не вычисляются оптимальные гиперпараметры).
   * предполагалось использовать 2 итерации: первая на полном датасете, чтобы определить, какие признаки важны (а вернее удостовериться, что всякие "интернет признаки" не важны). На второй, обучение на датасете, из которого удалены не важные признаки. По факту уже на первой итерации было достигнуто значение целевой метрики AUC-ROC ~ 0.94 при пороге в 0.85. Вторую итерацию в итоге проводил в точечном режиме на гиперпараметрах, взятых из первой итерации. Судя по замерам, удаление признаков \[`Dependents`, `SeniorCitizen`, `is_male`, `PaperlessBilling`, `TechSupport`, `StreamingMovies`, `OnlineSecurity`, `is_optic`, `Partner`, `StreamingTV`, `DeviceProtection`, `OnlineBackup`\] пракически не сказывается на целевой метрике, зато в 4-5 раз повышает скорость предсказаний.
   * На первой итерации лучшие модели показали следующие результаты:
      * LogisticRegression: оптимальные гиперпараметры: \['C': 10, 'max_iter': 100\], AUC-ROC: 0.790, время предсказания 0.111 сек.
      * RandomForestClassifier: оптимальные гиперпараметры: \[max_depth': 11, 'min_samples_leaf': 6, 'n_estimators': 40\], AUC-ROC: 0.907, время предсказания 1.677 сек.
      * CatBoostClassifier: оптимальные гиперпараметры: дефолтные (`GridSearchCV` не умеет их перебирать), AUC-ROC: 0.940, время предсказания 1.741 сек.
      * DummyClassifier: гиперпараметры отсутствуют. AUC-ROC: 0.5, время предсказания 1.740 сек.
   * На второй итерации, при тех же гиперпараметрах, достигнуты следующие результаты:
      * LogisticRegression: AUC-ROC: 0.779, время предсказания 0.097 сек.
      * RandomForestClassifier: AUC-ROC: 0.911, время предсказания 0.443 сек.
      * CatBoostClassifier: AUC-ROC: 0.940, время предсказания 0.381 сек.
   * Для каждой модели вывдилась гистограмма значимости признаков конкретно под эту модель.
