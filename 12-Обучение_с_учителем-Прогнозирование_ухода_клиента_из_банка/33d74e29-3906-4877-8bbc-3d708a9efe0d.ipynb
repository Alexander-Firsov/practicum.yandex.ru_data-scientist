```python
random_state = 12345 #v2
def prnd(i): return (i * 6723657243463463623) % 4294967291 # –≥–µ–Ω–µ—Ä–∞—Ç–æ—Ä –ø—Å–µ–≤–¥–æ—Å–ª—É—á–∞–π–Ω—ã—Ö —á–∏—Å–µ–ª
numeric_features = ['credit_score', 'age', 'tenure', 'balance', 'num_of_products',
                    'has_cr_card', 'is_active_member', 'estimated_salary']
categorical_features = ['geography', 'gender']
numeric_transformer = Pipeline(steps=[('scaler', StandardScaler()), ('imputer', KNNImputer())])
categorical_transformer = Pipeline(steps=[('encoder', OrdinalEncoder()), ('scaler', StandardScaler())])
preprocessor = ColumnTransformer(transformers=[
    ('num', numeric_transformer, numeric_features),
    ('cat', categorical_transformer, categorical_features)
    ])

class_weight=None

def upsample_plugin(train): # –∑–∞–≥–ª—É—à–∫–∞
    return train

up_down_sample = upsample_plugin

#def get_metric(*, train, valid, max_depth, n_estimators, preprocessor): #v1
def get_metric(*, train, valid, max_depth, n_estimators, regressor): #v2
    '''
    –¢—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã–π –∏ –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω—ã–π –¥–∞—Ç–∞—Å–µ—Ç —Ä–∞–∑–±–∏–≤–∞—é—Ç—Å—è –Ω–∞ –ø—Ä–∏–∑–Ω–∞–∫–∏ –∏ —Ü–µ–ª–∏, –≤—ã–∑—ã–≤–∞–µ—Ç—Å—è –≤–Ω–µ—à–Ω–∏–π (–ø–∞—Ä–∞–º–µ—Ç—Ä)
    pipeline —Å –ø—Ä–µ–¥–ø–æ–¥–≥–æ—Ç–æ–≤–∫–æ–π –¥–∞–Ω–Ω—ã—Ö. –ó–∞—Ç–µ–º –Ω–∞ –ø–æ–ª—É—á–∏–≤—à–µ–º—Å—è —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–º –¥–∞—Ç–∞—Å–µ—Ç–µ –æ–±—É—á–∞–µ—Ç—Å—è
    –º–æ–¥–µ–ª—å —Å–ª—É—á–∞–π–Ω–æ–≥–æ –ª–µ—Å–∞, —Å –≤–Ω–µ—à–Ω–∏–º–∏ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º–∏. –î–µ–ª–∞—é—Ç—Å—è –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏—è –¥–ª—è —Ç—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω–æ–π –∏
    –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –º–æ–¥–µ–ª–∏. –ù–∞ –≤—ã—Ö–æ–¥–µ —Ñ—É–Ω–∫—Ü–∏–∏ –Ω–∞–±–æ—Ä –º–µ—Ç—Ä–∏–∫, –≤—ã—á–µ—Å–ª–µ–Ω–Ω—ã—Ö –Ω–∞ –æ—Å–Ω–æ–≤–µ –ø—Ä–µ–¥—Å–∫–∞–∑–∞–Ω–∏–π.
    '''
    time_start_fit = time.time()
    train = up_down_sample(train) # v2
    features_train, features_valid = train.iloc[:,:-1], valid.iloc[:,:-1]
    target_train, target_valid = train.iloc[:,-1], valid.iloc[:,-1]
    if regressor == 'forest': #v2
        regressor = Pipeline(steps=[('preprocessor', preprocessor), ('classifier', #v2
            RandomForestClassifier(max_depth=max_depth, n_estimators=n_estimators, #v2
                random_state=random_state, class_weight=class_weight))]) #v2
    elif regressor == 'tree': #v2
        regressor = Pipeline(steps=[('preprocessor', preprocessor), ('classifier', #v2
            DecisionTreeClassifier(max_depth=max_depth, #v2
            random_state=random_state, class_weight=class_weight))]) #v2
    regressor.fit(features_train, target_train.values.ravel())
    time_finish_fit = time.time()
    def get_metric_one(features_valid, target_valid):
        target_pred = regressor.predict(features_valid)
        f1=f1_score(target_valid, target_pred)
        probabilities_valid = regressor.predict_proba(features_valid)
        probabilities_one_valid = probabilities_valid[:, 1]
        auc_roc = roc_auc_score(target_valid, probabilities_one_valid)
        return f1, auc_roc
    f1_train, auc_roc_train = get_metric_one(features_train, target_train)
    f1_valid, auc_roc_valid = get_metric_one(features_valid, target_valid)
##    print(f'{max_depth_global=} {n_estimators_global=} {f1_valid=} {f1_train=}')
    return pd.Series({'f1_train': f1_train, 'auc_roc_train':auc_roc_train, 'f1_valid': f1_valid,
        'auc_roc_valid':auc_roc_valid, 'fit_time': time_finish_fit - time_start_fit,
        'predict_time': time.time() - time_finish_fit})

#def get_metric_for_cross_validation(*, data_train,max_depth, n_estimators, n_splits, i, preprocessor): #v1
def get_metric_for_cross_validation(*, data_train, max_depth, n_estimators, n_splits, i, regressor): #v2
    '''
    –¢—Ä–µ–Ω–∏—Ä–æ–≤–æ—á–Ω—ã–π –¥–∞—Ç–∞—Å–µ—Ç —Ä–∞–∑–±–∏–≤–∞–µ—Ç—Å—è –Ω–∞ —á–∞—Å—Ç–∏ –¥–ª—è –∫—Ä–æ—Å—Å-–≤–∞–ª–∏–¥–∞—Ü–∏–∏ (—Å–æ —Å—Ç—Ä–∞—Ç–∏—Ñ–∏–∫–∞—Ü–∏–µ–π –ø–æ —Ü–µ–ª–µ–≤–æ–º—É
    –ø—Ä–∏–∑–Ω–∞–∫—É). –î–ª—è –∫–∞–∂–¥–æ–≥–æ –Ω–∞–±–æ—Ä–∞ —Ñ–æ–ª–¥–æ–≤ —Å–Ω–∏–º–∞—é—Ç—Å—è –º–µ—Ç—Ä–∏–∫–∏. –î–∞–ª–µ–µ –∏–¥–µ—Ç —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ—Å–∫–∞—è –æ–±—Ä–∞–±–æ—Ç–∫–∞ –º–µ—Ç—Ä–∏–∫
    (min, max, mean, std). –ù–∞ –≤—ã—Ö–æ–¥–µ - –Ω–∞–±–æ—Ä —Å—Ç–∞—Ç–∏—Å—Ç–∏—á–µ–∫—Å–∫–∏ –æ–±—Ä–∞–±–æ—Ç–∞–Ω–Ω—ã—Ö –º–µ—Ç—Ä–∏–∫ + –∏—Å–ø–æ–ª—å–∑–æ–≤–∞–≤—à–∏–µ—Å—è
    –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä—ã –º–æ–¥–µ–ª–∏.
    '''
    elementary_metric = pd.DataFrame(columns=['f1_train', 'auc_roc_train', 'f1_valid',
        'auc_roc_valid', 'fit_time', 'predict_time'])
    #kf = StratifiedKFold(n_splits, shuffle=True, random_state=12345) #v1
    kf = StratifiedKFold(n_splits, shuffle=True, random_state=random_state) #v2
    for train_data, valid_data in kf.split(data_train.iloc[:,:-1], data_train.iloc[:,-1]):
        metric = get_metric(train=data_train.iloc[train_data], valid=data_train.iloc[valid_data],
            #max_depth=max_depth, n_estimators=n_estimators, preprocessor=preprocessor) #v1
            max_depth=max_depth, n_estimators=n_estimators, regressor=regressor) #v2
        elementary_metric.loc[len(elementary_metric)] = metric
    res_dict = {}
    for column in elementary_metric:
        res_dict[column+'_min'] = elementary_metric[column].min()
        res_dict[column+'_max'] = elementary_metric[column].max()
        res_dict[column+'_mean'] = elementary_metric[column].mean()
        res_dict[column+'_std'] = elementary_metric[column].std()
    res_dict['max_depth'] = max_depth
    res_dict['n_estimators'] = n_estimators
    res_dict['n_splits'] = n_splits
    res_dict['rnd'] = i
    return pd.Series(res_dict)

#def get_df_stat1(*, data, max_depth_int, n_estimators_int, file=None, n_splits=5, i, preprocessor): #v1
def get_df_stat1(*, data, max_depth_int, n_estimators_int, file=None, n_splits=5, i, regressor): #v2
    '''
    –ì–µ–Ω–µ—Ä–∞—Ü–∏—è –¥–∞—Ç–∞—Å–µ—Ç–∞ —Å –º–µ—Ç—Ä–∏–∫–∞–º–∏, –ø—Ä–∏ —Ä–∞–∑–Ω—ã—Ö –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–∞—Ö –∏ –≤–∞—Ä–∏–∞–Ω—Ç–∞—Ö —Ä–∞–∑–±–∏–µ–Ω–∏—è.
    –ü—Ä–∏ —É–∫–∞–∑–∞–Ω–∏–∏ –∏–º–µ–Ω–∏ —Ñ–∞–π–ª–∞, –≤ –Ω–µ–≥–æ –±—É–¥–µ—Ç —Å–æ—Ö—Ä–∞–Ω–µ–Ω –¥–∞—Ç–∞—Å–µ—Ç.
    '''
    max_depth_int[1] += 1 # –ø—Ä–æ—Å—Ç–æ –¥–ª—è —É–¥–æ–±—Å—Ç–≤–∞ –≤–æ—Å–ø—Ä–∏—è—Ç–∏—è. –ß—Ç–æ–±—ã [—ã, —ä] –æ–±–æ–∑–Ω–∞—á–∞–ª –æ—Ç—Ä–µ–∑–æ–∫, –∞ –Ω–µ
    n_estimators_int[1] += 1 # –ø–æ–ª—É–æ—Ç—Ä–µ–∑–æ–∫
    i[1] += 1
    df_stat = pd.DataFrame(columns=[
       'f1_train_min', 'f1_train_max', 'f1_train_mean', 'f1_train_std',
       'auc_roc_train_min', 'auc_roc_train_max', 'auc_roc_train_mean',
       'auc_roc_train_std', 'f1_valid_min', 'f1_valid_max', 'f1_valid_mean',
       'f1_valid_std', 'auc_roc_valid_min', 'auc_roc_valid_max',
       'auc_roc_valid_mean', 'auc_roc_valid_std', 'fit_time_min',
       'fit_time_max', 'fit_time_mean', 'fit_time_std', 'predict_time_min',
       'predict_time_max', 'predict_time_mean', 'predict_time_std',
       'max_depth', 'n_estimators', 'n_splits', 'rnd'])
    progress_bar = widgets.IntProgress(
        value=0,
        min=0,
        max=(max_depth_int[1]-max_depth_int[0])*math.ceil((n_estimators_int[1]-n_estimators_int[0])/5)*(i[1]-i[0]),
        description='–ó–∞–≤–µ—Ä—à–µ–Ω–æ:',
        bar_style='', # 'success', 'info', 'warning', 'danger' or ''
        style={'bar_color': 'maroon'},
        orientation='horizontal'
    )
    j=0
    time_start = time.time()
    display(f'–°—Ç–∞—Ä—Ç: {datetime.datetime.fromtimestamp(time_start)}')
    # —Ç–µ—Å—Ç–æ–≤—ã–π –∑–∞–º–µ—Ä –≤—Ä–µ–º–µ–Ω–∏ (–¥–ª—è –ø—Ä–æ–≥–Ω–æ–∑–∞ –≤—Ä–µ–º–µ–Ω–∏ –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏)
    #tmp = get_metric_for_cross_validation( #v1
    #    data_train=data, max_depth=max_depth_int[1], preprocessor=preprocessor, #v1
    #        n_estimators=n_estimators_int[1], n_splits=n_splits, i=i) #v1
    # —Å—Ç–∞—Ç–∏—á–µ—Å–∫–æ–µ —Ä–∞–∑–±–∏–µ–Ω–∏–µ
    train, test = train_test_split(data, test_size = 0.1, random_state = i[0], stratify=data.exited) #v2
    tmp = get_metric_for_cross_validation(regressor=regressor, data_train=train, max_depth=max_depth_int[1], #v2
            n_estimators=n_estimators_int[1], n_splits=n_splits, i=i[0]) #v2
    time_one_fit = time.time() - time_start
    time_full = time_one_fit * (progress_bar.max + 1) + time_start
    display(f'–û—Ä–∏–µ–Ω—Ç–∏—Ä–æ–≤–æ—á–Ω–æ–µ –≤—Ä–µ–º—è –∑–∞–≤–µ—Ä—à–µ–Ω–∏—è –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ –¥–∞—Ç–∞—Å–µ—Ç–∞: {datetime.datetime.fromtimestamp(time_full)}')
    display(progress_bar)
    for k in range(i[0],i[1]): # –ø—Ä–æ–±—É–µ–º i —Ä–∞–∑ –ø–æ —Ä–∞–∑–Ω–æ–º—É —Ä–∞–∑–±–∏—Ç—å –∏–∑–Ω–∞—á–∞–ª—å–Ω—ã–µ –¥–∞–Ω–Ω—ã–µ
        #train, test = train_test_split(data, test_size = 0.1, random_state = prnd(k), stratify=data.exited) #v1
        # train = up_down_sample(train) #v1
        for max_depth in range(max_depth_int[0],max_depth_int[1]):
            for n_estimators in range(n_estimators_int[0], n_estimators_int[1],5):
                #df_stat.loc[len(df_stat)] = get_metric_for_cross_validation(preprocessor=preprocessor, #v1
                df_stat.loc[len(df_stat)] = get_metric_for_cross_validation(regressor=regressor, #v2
                    data_train=train, max_depth=max_depth, n_estimators=n_estimators, n_splits=n_splits, i=k)
                j += 1
                progress_bar.value = j
    if file != None:
        df_stat.to_csv(file, index=False)
        display(f'–î–∞—Ç–∞—Ñ—Ä–µ–π–º —Å–æ—Ö—Ä–∞–Ω–µ–Ω –≤ —Ñ–∞–π–ª {file}')
    display(f'–ó–∞–≤–µ—Ä—à–µ–Ω–∏–µ: {datetime.datetime.fromtimestamp(time.time())}')
    return df_stat

def two_colimn_cross_count2(data,column1,column2, values, aggfunc):
    '''
    –í—Å–ø–æ–º–æ–≥–∞—Ç–µ–ª—å–Ω–∞—è —Ñ—É–Ω–∫—Ü–∏—è, –¥–ª—è –ø–æ—Å—Ç—Ä–æ–µ–Ω–∏—è —Ç–µ–ø–ª–æ–≤–æ–π –∫–∞—Ä—Ç—ã –ø–æ 2 –ø–∞—Ä–∞–º–µ—Ç—Ä–∞–º.
    '''
    result=pd.DataFrame(columns=sorted(data[column1].unique()), index=sorted(data[column2].unique()))
    result=result.fillna(0)
    pivot_table=data.pivot_table(index=[column1,column2], values=values, aggfunc = aggfunc).reset_index()
    for i,v in pivot_table.iterrows():
        result.loc[v[column2],v[column1]] = v[values]
    return result

def df_stat_visualisation(*, data, f1_mean_min, f1_std_max):
    '''
    –í–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è –¥–∞—Ç–∞—Å–µ—Ç–∞ —Å –º–µ—Ç—Ä–∏–∫–∞–º–∏
    '''
    tmp = data[data.f1_valid_mean >= f1_mean_min[1]]
    window=[tmp.max_depth.min(),tmp.n_estimators.min()]
    figure, grid = plt.figure(figsize=(25, 20)), GridSpec(20, 3)
    figure.add_subplot(grid[0:2, :])
    figure.add_subplot(grid[3:5, :])
    figure.add_subplot(grid[6:8, :])
    figure.add_subplot(grid[9:11, :])
    figure.add_subplot(grid[12:15, 0])
    figure.add_subplot(grid[12:15, 1])
    figure.add_subplot(grid[12:15, 2])
    figure.add_subplot(grid[16:19, 0])
    figure.add_subplot(grid[16:19, 1])
    figure.add_subplot(grid[16:19, 2])
    
    hh_f1_mean = two_colimn_cross_count2(data, 'n_estimators',  'max_depth','f1_valid_mean', 'mean')
    hh_f1_std=two_colimn_cross_count2(data, 'n_estimators',  'max_depth','f1_valid_std', 'mean')
    tmp = sns.heatmap(hh_f1_mean, ax=figure.axes[0], vmin=f1_mean_min[0])
    figure.axes[0].set (ylabel='max_depth',
        xlabel='–∑–Ω–∞—á–µ–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ n_estimators',
        title=f'—Ç–µ–ø–ª–æ–≤–∞—è –∫–∞—Ä—Ç–∞ mean f1 –¥–ª—è –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏ (–ø–æ—Ä–æ–≥ {f1_mean_min[0]})');
    
    tmp = sns.heatmap(hh_f1_mean.loc[window[0]:,window[1]:],ax=figure.axes[1], vmin=f1_mean_min[1])
    figure.axes[1].set (ylabel='max_depth',
        xlabel='–∑–Ω–∞—á–µ–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ n_estimators',
        title=f'–≤—ã—Ä–µ–∑–∫–∞ –∏–∑ —Ç–µ–ø–ª–æ–≤–æ–π –∫–∞—Ä—Ç—ã (window) mean f1 –¥–ª—è –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏ (–ø–æ—Ä–æ–≥ {f1_mean_min[1]})');
    
#     tmp = sns.heatmap(hh_f1_std,ax=figure.axes[2], vmax=f1_std_max[0])
#     figure.axes[2].set (ylabel='max_depth',
#         xlabel='–∑–Ω–∞—á–µ–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ n_estimators',
#         title=f'—Ç–µ–ø–ª–æ–≤–∞—è –∫–∞—Ä—Ç–∞ std f1 –¥–ª—è –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏ (–ø–æ—Ä–æ–≥ {f1_std_max[0]})');
    
#     tmp = sns.heatmap(hh_f1_std.loc[window[0]:,window[1]:],ax=figure.axes[3], vmax=f1_std_max[1])
#     figure.axes[3].set (ylabel='max_depth',
#         xlabel='–∑–Ω–∞—á–µ–Ω–∏–µ –ø–∞—Ä–∞–º–µ—Ç—Ä–∞ n_estimators',
#         title=f'–≤—ã—Ä–µ–∑–∫–∞ –∏–∑ —Ç–µ–ø–ª–æ–≤–æ–π –∫–∞—Ä—Ç—ã (window) std f1 –¥–ª—è –≤–∞–ª–∏–¥–∞—Ü–∏–æ–Ω–Ω–æ–π –≤—ã–±–æ—Ä–∫–∏ (–ø–æ—Ä–æ–≥ {f1_std_max[1]})');
    
    ax=data.plot.scatter(x='f1_valid_mean',
            y='auc_roc_valid_mean', ax=figure.axes[4])
    figure.axes[4].set (ylabel='mean auc_roc',
        xlabel='mean f1',
        title='–¥–∏–∞–≥—Ä–∞–º–º–∞ —Ä–∞—Å—Å–µ—è–Ω–∏—è f1 –∏ auc_roc (full)');
    
    ax=data.plot.scatter(x='f1_valid_mean',
            y='fit_time_mean', ax=figure.axes[5])
    figure.axes[5].set (ylabel='fit_time, —Å–µ–∫',
        xlabel='mean f1',
        title='–¥–∏–∞–≥—Ä–∞–º–º–∞ —Ä–∞—Å—Å–µ—è–Ω–∏—è f1 –∏ fit_time (full)');
    
    ax=data.plot.scatter(x='f1_valid_mean',
            y='predict_time_mean', ax=figure.axes[6])
    figure.axes[6].set (ylabel='predict_time, —Å–µ–∫',
        xlabel='mean f1',
        title='–¥–∏–∞–≥—Ä–∞–º–º–∞ —Ä–∞—Å—Å–µ—è–Ω–∏—è f1 –∏ predict_time (full)');
    
    ax=data[data.f1_valid_mean >= f1_mean_min[1]].plot.scatter(x='f1_valid_mean',
            y='auc_roc_valid_mean', ax=figure.axes[7])
    figure.axes[7].set (ylabel='mean auc_roc',
        xlabel='mean f1',
        title='–¥–∏–∞–≥—Ä–∞–º–º–∞ —Ä–∞—Å—Å–µ—è–Ω–∏—è f1 –∏ auc_roc (window)');
    
    ax=data[data.f1_valid_mean >= f1_mean_min[1]].plot.scatter(x='f1_valid_mean',
            y='fit_time_mean', ax=figure.axes[8])
    figure.axes[8].set (ylabel='fit_time, —Å–µ–∫',
        xlabel='mean f1',
        title='–¥–∏–∞–≥—Ä–∞–º–º–∞ —Ä–∞—Å—Å–µ—è–Ω–∏—è f1 –∏ fit_time (window)');
    
    ax=data[data.f1_valid_mean >= f1_mean_min[1]].plot.scatter(x='f1_valid_mean',
            y='predict_time_mean', ax=figure.axes[9])
    figure.axes[9].set (ylabel='predict_time, —Å–µ–∫',
        xlabel='mean f1',
        title='–¥–∏–∞–≥—Ä–∞–º–º–∞ —Ä–∞—Å—Å–µ—è–Ω–∏—è f1 –∏ predict_time (window)');
    
def get_df_stat2(*, data, max_depth_int, n_estimators_int, file=None, n_splits=5, i,
    #preprocessor, need_to_be_compiled, f1_mean_min, f1_std_max): #v1
    regressor, need_to_be_compiled, f1_mean_min, f1_std_max): #v2
    '''
    –ì–µ–Ω–µ—Ä–∞—Ü–∏—è/–∑–∞–≥—Ä—É–∑–∫–∞ –¥–∞—Ç–∞—Å–µ—Ç–∞ —Å –º–µ—Ç—Ä–∏–∫–∞–º–∏, –ø–æ–ª—É—á–µ–Ω–Ω—ã–º–∏ –ø—Ä–∏ —Ä–∞–∑–Ω—ã—Ö –≥–∏–ø–µ—Ä–ø–∞—Ä–∞–º–µ—Ç—Ä–∞—Ö –º–æ–¥–µ–ª–∏ –∏
    —Ä–∞–∑–±–∏–µ–Ω–∏—è—Ö. –ü–æ—Å–ª–µ –ø–æ–ª—É—á–µ–Ω–∏—è –¥–∞—Ç–∞—Å–µ—Ç–∞ –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç—Å—è –µ–≥–æ –≤–∏–∑—É–∞–ª–∏–∑–∞—Ü–∏—è.
    –ï—Å–ª–∏ –ø—Ä–æ–∏–∑–≤–æ–¥–∏—Ç—Å—è –∑–∞–≥—Ä—É–∑–∫–∞ —Å–æ—Ö—Ä–∞–Ω–µ–Ω–Ω–æ–≥–æ –¥–∞—Ç–∞—Å–µ—Ç–∞ (—É—Å—Ç–∞–Ω–æ–≤–∫–∞ –∞—Ä–≥—É–º–µ–Ω—Ç–∞ need_to_be_compiled = True),
    —Ç–æ –ø–µ—Ä–µ–º–µ–Ω–Ω—ã–µ –≥–µ–Ω–µ—Ä–∞—Ü–∏–∏ –¥–∞—Ç–∞—Å–µ—Ç–∞ –∏–≥–Ω–æ—Ä–∏—Ä—É—é—Ç—Å—è.
    '''
    if not need_to_be_compiled:
        try:
            df_stat = pd.read_csv(file)
        except:
            try:
                df_stat = pd.read_csv('https://gifara.ru/ds/'+file)
            except:
                need_to_be_compiled = True
    if need_to_be_compiled:
        df_stat = get_df_stat1(data=data, max_depth_int=max_depth_int, n_estimators_int=n_estimators_int,
            #n_splits=8, i=i, preprocessor=preprocessor, file=file) #v1
            n_splits=n_splits, i=i, regressor=regressor, file=file) #v2
    df_stat_visualisation(data=df_stat, f1_mean_min = f1_mean_min, f1_std_max = f1_std_max)
    return df_stat
```
